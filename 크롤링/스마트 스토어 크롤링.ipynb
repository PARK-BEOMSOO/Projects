```python
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome('chromedriver',options=options)
time.sleep(3.0)
```

```python
url = 'https://smartstore.naver.com/aunatural/products/319810583'
html = urlopen(url)
soup = BeautifulSoup(html, "html.parser", from_encoding='utf-8')

driver.get(url)
time.sleep(3) 

driver.find_element_by_css_selector('body').send_keys(Keys.CONTROL + "t")

review = driver.find_element_by_css_selector('span._3HJHJjSrNK')
review.click()
time.sleep(3)

review_id_infos= []
review_content_infos = []
 
etc = 1
while etc:
    try:
        for x in range(1,11):
            try : 
                for i in range(1,21):
                        #try: 
                            print("======================================================")

                            html_source = driver.page_source 

                            html_soup = BeautifulSoup(html_source, 'lxml') 

                            # 1) 채용 직무 job_name_infos
                            review_id = html_soup.find_all('strong', {'class':'_3QDEeS6NLn'})[i].text
                            print(review_id)

                            review_id_infos.append(review_id)

                            # 1) 채용 직무 job_name_infos
                            div = html_soup.find_all('div', {'class':'YEtwtZFLDz'})[i]
                            review_content = div.find('span',{'class':'_3QDEeS6NLn'}).text
                            print(review_content)

                            review_content_infos.append(review_content)

                            review_dic = {"아이디":review_id_infos, "리뷰상세내용":review_content_infos} 

                            review_df = pd.DataFrame(review_dic)

                        #except:
                            print("pass2")

                page = driver.find_all('a', {'class': 'UWN4IvaQza N=a:rvs.page'})[x]
                page.click()
                time.sleep(3)

            except:
                print("pass1")
        total =  driver.find_all('a', {'class': 'fAUKm1ewwo _2Ar8-aEUTq'})[0]
        total.click()
        time.sleep(3)
    except:
        break
        
review_df.to_csv(r'C:\Users\qkrqj\workspace\JupyterNotebook\csv\스마트스토어_리뷰.csv', encoding='UTF-8-SIG')
```
